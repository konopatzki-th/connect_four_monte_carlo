```python
from src.players import random_player, heuristic_player, intelligent_player
from src.simulation import monte_carlo

players = {
    "4 Years": random_player,
    "7 Years": heuristic_player,
    "9 Years": lambda b, p: intelligent_player(b, p, depth=1),
    "11 Years": lambda b, p: intelligent_player(b, p, depth=2)
}

results = monte_carlo(players)

for (p1, p2), data in results.items():
    p1_prob, p2_prob, draw_prob = data["probabilities"]
    se1, se2, sed = data["SE"]
    q1, q2, qd = data["quotes"]
    print(f"\n{p1} vs {p2} ({data['games']} games)")
    print(f"{p1}: {p1_prob*100:.1f}% ± {se1*100:.1f}% → Quote ≈ {q1:.2f}")
    print(f"{p2}: {p2_prob*100:.1f}% ± {se2*100:.1f}% → Quote ≈ {q2:.2f}")
    print(f"Draw: {draw_prob*100:.1f}% ± {sed*100:.1f}% → Quote ≈ {qd:.2f}")
```
